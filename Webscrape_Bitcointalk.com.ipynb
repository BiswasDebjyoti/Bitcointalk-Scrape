# Websraping using requests and beautifulsoup and regex to find numbers

### Importing Stuff

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup as bs
```

## URL and Sending Reuest

```python
url="https://bitcointalk.org/index.php?board=121.0"
r=requests.get(url)
print r
# Response 200 means request is accepted
```

# Using BeautifulSoup to Beautify

```python
soup = bs(r.content,"lxml")
print soup
```

# Finding the div tag of our need

```python
data = soup.find_all("div",{"id":"bodyarea"})

print data
```

```python
number=3
names=[]
links=[]
count =0
import xlwt
wb=xlwt.Workbook()
ws = wb.add_sheet("sheet 1")
for item in data:
    je=item.contents[5].contents[1]
    for _,task in enumerate(je):
        try:
            bub= item.contents[5].contents[1].contents[number].contents[5].contents[1].contents[0]
            #print bub.text
            #print bub['href']
            ws.write(count,0,bub.text)
            ws.write(count,1,bub['href'])
            names.append(bub.text)
            links.append(bub['href'])
        except IndexError:
            try:
                bub= item.contents[5].contents[1].contents[number].contents[5].contents[3].contents[0]
                ws.write(count,0,bub.text)
                ws.write(count,1,bub['href'])
                #print bub.text
                #print bub['href']
            except:
                break
        number+=2
        count+=1
    wb.save("data.xls") 
```

## Creating Dict of Topic to link

```python
certain_dict={'Topics':names,'Links':links}
print certain_dict
```

## creating pandas dataframe

```python
df=pd.DataFrame(certain_dict)
print df
```

```python
df.to_csv("link.csv")
```
